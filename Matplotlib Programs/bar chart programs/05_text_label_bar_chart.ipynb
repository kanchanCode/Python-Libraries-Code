```python
'''
@Author: Kanchan
@Date: 2022-01-05
@Last Modified by: Kanchan
@Last Modified time: 2022-01-05
@Title : Write a Python programming to display a bar chart of the popularity of programming
Languages. Attach a text label above each bar displaying its popularity (float value).
'''
import matplotlib.pyplot as plt

def addlabel(x,y):
    box_style=dict(boxstyle='round', facecolor='wheat', alpha=0.5)

    for i in range(len(x)):
        plt.text(i,y[i],y[i],ha='center',bbox=box_style)
        #plt.txt(x,y,text,text property)

if __name__=='__main__':
    
    x=['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    
    plt.xlabel("Languages")
    plt.ylabel("Popularity")
    plt.title("Popularity of programming language in Oct2017")
    
    plt.bar(x,popularity,color=['blue','pink','cyan','maroon','orange','red'],edgecolor='yellow')
    addlabel(x,popularity)

    plt.show()
```
