```python
'''
@Author: Kanchan
@Date: 2022-01-04
@Last Modified by: Kanchan
@Last Modified time: 2022-01-04
@Title : Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.
'''
from matplotlib import pyplot as plt
with open("03_file.txt")as f:
    data=f.read()
data=data.split('\n')
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]
print(x)
print(y)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Data analysis')
plt.show()

```
