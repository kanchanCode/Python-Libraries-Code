```python
'''
@Author: Kanchan
@Date: 2022-01-04
@Last Modified by: Kanchan
@Last Modified time: 2022-01-04
@Title : Write a Python program to create multiple plots
'''
import matplotlib.pyplot as plt
import numpy as np

#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,9])

plt.subplot(1,3,1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,3,2)
plt.plot(x,y)

#plot 3

x = [1, 12, 3, 9]
y = [1, 4, 9, 16]
labels = ['Data1', 'Data2', 'Data3', 'Data4']
plt.subplot(1,3,3)
plt.plot(x, y)
plt.xticks(x, labels, rotation ='vertical')
  
plt.margins(0.2)
plt.subplots_adjust(bottom = 0.15)
  
plt.title('Multiple plots')
plt.show()



```
