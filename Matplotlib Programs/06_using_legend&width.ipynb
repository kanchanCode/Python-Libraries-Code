```python
'''
@Author: Kanchan
@Date: 2022-01-04
@Last Modified by: Kanchan
@Last Modified time: 2022-01-04
@Title : Write a Python program to plot two or more lines with legends, different widths and colors.

'''
from matplotlib import pyplot as plt

x=[20,40,60,80]
y=[10,30,60,70]

x2=[10,30,50,70]
y2=[30,50,40,80]

plt.plot(x,y,color='red',linewidth=3,label="line A")
plt.plot(x2,y2,color='green',linewidth=4,label="line B")

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two lines on same plot with legends')
plt.legend()
plt.show()
```
