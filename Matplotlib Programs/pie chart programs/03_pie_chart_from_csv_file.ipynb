```python
'''
@Author: Kanchan
@Date: 2022-01-05
@Last Modified by: Kanchan
@Last Modified time: 2022-01-05
@Title : Write a Python programming to create a pie chart of gold medal achievements of five most
successful countries in 2016 Summer Olympics. Read the data from a csv file.
Data:
medal.csv
country,gold_medal
United States,46
Great Britain,27
China,26
Russia,19
Germany,17

'''

import matplotlib.pyplot as plt
import pandas as pd
df =  pd.read_csv('03medal.csv')

country_data = df["country"]
medal_data = df["gold_medal"]

colors=['red','blue','green','yellow','cyan','maroon']
explode=(0.1,0,0,0,0)
plt.pie(medal_data, labels=country_data, explode=explode, colors=colors,
autopct='%.2f%%', shadow=True, startangle=140)
#autopct -> to show percentages
plt.title("Medal's tally ")
plt.show()
```
