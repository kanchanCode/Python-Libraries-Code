```python
'''
@Author: Kanchan
@Date: 2022-01-05
@Last Modified by: Kanchan
@Last Modified time: 2022-01-05
@Title : Write a Python programming to create a pie chart of the popularity of programming
Languages.
Data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
'''

import matplotlib.pyplot as plt
# Data to plot
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

colors=['red','blue','green','yellow','cyan','maroon']
explode=(0.1,0,0,0,0,0)

plt.pie(popuratity, explode=explode, labels=languages, colors=colors,
autopct='%.2f%%', shadow=True, startangle=140)
#autopct -> to show percentages


plt.axis('equal')
plt.show()
```
