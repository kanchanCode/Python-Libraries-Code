```python
'''
@Author: Kanchan
@Date: 2022-01-06
@Last Modified by: Kanchan
@Last Modified time: 2022-01-06
@Title : Write a Python program to draw a scatter plot with empty circles taking a random distribution
in X and Y and plotted against each other
'''

import matplotlib.pyplot as plt
from pylab import randn

#The numpy.random.randn() function creates an array of specified shape and fills it with 
# random values as per standard normal distribution.
#The randn function generates arrays of random numbers whose elements are normally distributed with mean 0, variance , and standard deviation.
#normal distribution=It is the distribution that occurs when a normal random variable has a mean of zero and a standard deviation of one
x=randn(100)
y=randn(100)
#print(x)
#print(y)
plt.scatter(x,y,fc='none',edgecolor='b')
#fc=facecolors
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
