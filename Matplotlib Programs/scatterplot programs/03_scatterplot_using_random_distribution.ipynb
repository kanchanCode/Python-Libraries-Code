```python
'''
@Author: Kanchan
@Date: 2022-01-06
@Last Modified by: Kanchan
@Last Modified time: 2022-01-06
@Title : Write a Python program to draw a scatter plot using random distributions to generate balls of
different sizes
'''

import math
import matplotlib.pyplot as plt
import random

n=25

x=[random.triangular() for i in range(n)]
#x=[random.random() for i in range(n)]
print(x)
#The triangular() method returns a random floating number

y=[random.gauss(0.5,0.25) for i in range(n)]
# can use normalvariate() also
print(y)
#gauss() is an inbuilt method of the random module. It is used to return a random floating point number with gaussian distribution.
#random.gauss(mu, sigma)
#mu -> mean ->defines the peak of the curve in normal distribution
#sigma -> std deviation-> defines the width of the curve in normal distribution

#sizes=[math.pi*random.randint(5,15)**2 for i in range(n)]
sizes=[random.randint(50,500) for i in range(n)]
print(sizes)

plt.figure()
#figure() is container for all types of plots
plt.scatter(x,y,s=sizes,color='g',alpha=0.75)
plt.axis('equal')
#it creates a square plot where the ranges for both axes occupy are equal to the length in plot.
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
