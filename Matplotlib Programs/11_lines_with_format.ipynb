```python
'''
@Author: Kanchan
@Date: 2022-01-04
@Last Modified by: Kanchan
@Last Modified time: 2022-01-04
@Title : Write a Python program to plot several lines with different format styles in one command
using arrays.

'''
import pylab as pl

time_=pl.arange(0,5.0,0.3)

pl.plot(time_,time_,'gs',time_,time_**2,'r^', time_,time_**3,'b*')
pl.xlabel('x-axis(Time in ms)')
pl.ylabel('y-axis(Rise in quantity)')
pl.title('Data analysis')
pl.show()
```
