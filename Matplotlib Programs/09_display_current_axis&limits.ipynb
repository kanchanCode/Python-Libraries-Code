```python
'''
@Author: Kanchan
@Date: 2022-01-04
@Last Modified by: Kanchan
@Last Modified time: 2022-01-04
@Title : Write a Python program to display the current axis limits values and set new axis values.
'''
from matplotlib import pyplot as plt

x=[20,40,60,80]
y=[10,30,60,70]

plt.plot(x,y,color='blue',linestyle='dotted',linewidth=2,marker='o',markerfacecolor='red',markersize=5,label="line A")

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Data analysis')
print("Old axis",plt.axis())
plt.axis([0,100,0,200])
print("New axis",plt.axis())
plt.show()
```
