# SEABORN
1. Data visualization & based on matplotlib
2. Allows the creation of statistical graphic

Functionality
1. supports multi-plots grid
2. comparison b/w multiple variables
3. do linear regression possible

Matplotlib v/s Seaborn

1. Matplotlib settings are difficult to figure out .Seaborn comes with numerous customized themes and high-level interfaces

2. Matplotlib dosen't serve well when it comes to dealing with DataFrames, while Seaborn func. actually work on DataFrame

# Seaborn Functions



```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

df= sns.load_dataset('titanic')
sns.barplot(x = "sex", y = "survived", hue = "class", data = df)
plt.show()


```

```python
df= sns.load_dataset('flights')
sns.relplot(x = "passengers", y = "month", hue = "year", data = df)
plt.show()
```

```python
df= sns.load_dataset('tips')
sns.relplot(x = "time", y = "tip", data = df, kind="line")
```

```python
sns.catplot(x = "day", y = "total_bill", data = df)
```

```python
sns.catplot(x = "day", y = "total_bill", data = df,kind="violin")
```

```python
sns.catplot(x = "day", y = "total_bill", data = df,kind="boxen")
```

```python
df=sns.load_dataset("iris")
b=sns.FacetGrid(df,col="species")
b.map(plt.hist,"sepal_length")
```

```python
df=sns.load_dataset("flights")
b=sns.PairGrid(df)
b.map(plt.scatter)
```

```python
sns.set(style="dark")
df=sns.load_dataset("flights")
b=sns.PairGrid(df)
b.map(plt.scatter)
```

```python
sns.set(style="white",color_codes=True)
df= sns.load_dataset('tips')
sns.boxplot(x = "day", y = "total_bill", data = df)
```

```python
sns.set(style="white",color_codes=True)
df= sns.load_dataset('tips')
sns.boxplot(x = "day", y = "total_bill", data = df)
sns.despine(offset=10,trim=True)
```
