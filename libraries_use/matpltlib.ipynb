```python
from matplotlib import pyplot as plt
from matplotlib import style as stl
stl.use('ggplot') 
#ggplot2 is a plotting package that provides helpful commands to create complex plots from data in a data frame. 
x=[1,5,8]
y=[23,45,78]
x2=[3,5,9]
y2=[30,45,80]
plt.title("Data analysis")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y,label="first",linewidth=3)
plt.plot(x2,y2,label="second",linewidth=3)
plt.scatter(x,y)
plt.scatter(x2,y2)

plt.legend()
plt.grid(True,color='k') #for grid  k is for black
plt.show()
```

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
sleeping=[8,7,6,8,10]
eating=[2,3,4,3,2]
working=[7,4,6,5,3]

plt.plot([],[],color='m',label='Sleeping',linewidth=5)
plt.plot([],[],color='c',label='Eating',linewidth=5)
plt.plot([],[],color='r',label='Working',linewidth=5)

plt.stackplot(days,sleeping,eating,working,colors=['m','c','r'])

plt.title("Data analysis through Area plot")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

slices=[7,2,2]
activities=['sleeping','eating','working']
colrs=['c','m','r']

plt.pie(slices, labels=activities,colors=colrs,
startangle=90,shadow=True,explode=(0.1,0,0),autopct='%1.1f%%')
#autopct -> to show percentages
plt.title("Data analysis through Pie plot")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(221)
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(222)
plt.plot(t2,np.cos(2*np.pi*t2))

plt.show()
```
