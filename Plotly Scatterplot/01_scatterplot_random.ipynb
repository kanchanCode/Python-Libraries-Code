```python
'''
@Author: Kanchan
@Date: 2022-01-07
@Last Modified by: Kanchan
@Last Modified time: 2022-01-07
@Title :Write a program to draw a scatter plot for random 1000 x and y coordinates
'''

import pandas as pd
import plotly.express as px
from pylab import randn

x=randn(1000)
y=randn(1000)

df=pd.DataFrame({'x-axis':x,'y-axis':y})

fig=px.scatter(df,x='x-axis',y='y-axis',width=700,height=500,title='Scatter Plot')
fig.show()
```
