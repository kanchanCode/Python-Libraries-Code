```python
'''
@Author: Kanchan
@Date: 2022-01-07
@Last Modified by: Kanchan
@Last Modified time: 2022-01-07
@Title :Write a program to draw line and scatter plots for random 100 x and y coordinates
'''

import pandas as pd
import plotly.express as px
from pylab import randn

x=randn(100)
y=randn(100)

df=pd.DataFrame({'x-axis':x,'y-axis':y})
fig=px.line(df,x='x-axis',y='y-axis',width=700,height=500,title="Line Plot")
fig.show()
```

```python
fig=px.scatter(df,x='x-axis',y='y-axis',width=700,height=500,title='Scatter Plot')
fig.show()
```
